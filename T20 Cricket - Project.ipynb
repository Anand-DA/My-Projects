```python
import pandas as pd
import json

```

```python
with open("t20_wc_match_results.json") as f:
    data = json.load(f)
    
data

data[0]['matchSummary']
```

```python
df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.shape
```

```python
df_match.rename({'scorecard': 'match_id'}, axis = 1, inplace = True)
```

```python
match_id_dict = {}

for index, row in df_match.iterrows():
    
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    
    match_id_dict[key1] = row['match_id']
    match_id_dict[key2] = row['match_id']
    
match_id_dict
```

```python
df_match.head()
```

```python
match_id_dict["Namibia Vs Sri Lanka"]
```

```python
df_match.to_csv('cric_match_summary.csv', index = False)
```

**Batting Summary**

<h4 style="color:blue">(2) Process Batting Summary</h4>

```python
with open("t20_wc_batting_summary.json") as f:
    
    data = json.load(f)
    
    all_records = []
    
    for rec in data:
        
        all_records.extend(rec['battingSummary'])
    
df_batting = pd.DataFrame(all_records)
df_batting.head()
```

```python
df_batting["out/not_out"] = df_batting.dismissal.apply(lambda x: "out" if len(x) > 0 else "not_out")
df_batting.head(11)
```

```python
df_batting.drop(columns = ['dismissal'], inplace = True)
df_batting.head()
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€', ''))
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('/â€', ''))
```

```python
df_batting.head(30)
```

```python
df_batting['match_id'] = df_batting['match'].map(match_id_dict)
df_batting
```

```python
df_batting.to_csv('cric_batting_summary.csv', index = False)
```

<h4 style="color:blue">(3) Process Bowling Summary</h4>

```python
with open('t20_wc_bowling_summary.json') as f:
    
    data = json.load(f)
    
    all_records = []
    
    for rec in data:
        
        all_records.extend(rec['bowlingSummary'])
all_records[:2]
```

```python
df_bowling = pd.DataFrame(all_records)
print(df_bowling.shape)
df_bowling.head()
```

```python
df_bowling['match_id'] = df_bowling['match'].map(match_id_dict)
df_bowling.head()
```

```python
df_bowling.to_csv('cric_bowling_summary.csv', index = False)
```

<h4 style="color:blue">(4) Process Players Summary</h4>

```python
with open('t20_wc_player_info.json') as f:
    
    data  = json.load(f)

df_players = pd.DataFrame(data)
df_players.tail(100)
```

```python
df_players['name'] = df_players['name'].apply(lambda x: x.replace('â€', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('†', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('\xa0', ''))
df_players.head(10)
```

```python
df_players[df_players['team'] == 'India']
```

```python
df_players.to_csv('cric_players_summary.csv', index = False)
```
